```python
#Exercício 2
nome1 = input("Digite o 1 \n nome: \n")
id1 =  float(input(f"Digite a altura do \n{nome1}"))
nome2 = input("Digite o 2 nome \n")
id2 = float(input(f"Digite a altura \n {nome2}"))
if (id1> id2):
    print(f"{nome1} é mais alto que \n{nome2}")
elif(id2> id1):
    print(f"{nome2} é mais alto que \n {nome1}")
else:
    print(f"{nome1} e {nome2} tem a mesma altura!")
```
